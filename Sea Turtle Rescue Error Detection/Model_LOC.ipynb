```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_clean = pd.read_csv("test_data _cleaned.csv", encoding='cp1252')
df_dirty = pd.read_csv("test_data_dirty.csv", encoding='cp1252')
df_dirty.head()

```

```python
#pd.set_option('display.max_columns', 30)

def get_stacked(df_clean):
    df_clean_stacked = pd.DataFrame()
    count1 = 0
    x=[]
    y=[]
    for i in range(0, len(df_clean)):
        for ix in range(1,26):
            x.append(str(df_clean.iloc[count1,0])+' x '+str(df_clean.columns.values[ix]))
            y.append(df_clean.iloc[count1,ix])
        count1+=1
    df1 = pd.DataFrame({"ID": x})  
    df1.index.names = ['index']
    df2 = pd.DataFrame({"Value": y}) 
    
    df2.index.names = ['index']
    df_clean_stacked = pd.merge(df1, df2, on='index')
    return df_clean_stacked      
```

```python
df_dirtyStacked = get_stacked(df_dirty)
df_cleanStacked = get_stacked(df_clean)
df_dirtyStacked.head()
```

```python
def get_label(df_d, df_c):
    df_clean_labeled = pd.DataFrame()
    new_labels = []
    
    for i in range(0, len(df_d)):
        if df_d.iloc[i,0] == df_c.iloc[i,0] and df_d.iloc[i,1] == df_c.iloc[i,1] or (pd.isnull(df_d.iloc[i,1]) == True and pd.isnull(df_c.iloc[i,1]) == True):
            new_labels.append(0)
        else:
            new_labels.append(1)
    
    df1 = pd.DataFrame(df_d.ID)  
    df1.index.names = ['index']
    df2 = pd.DataFrame({"error": new_labels})  
    df2.index.names = ['index']
    df3 = pd.merge(df1, df2, on='index')
    df_clean_labeled = df3.rename_axis(None)
    pd.options.display.float_format = '{:,.0f}'.format
    return df_clean_labeled
```

```python
labeled_data = get_label(df_dirtyStacked, df_cleanStacked)
labeled_data
```

```python
Y = labeled_data['error']
X = labeled_data['ID']
X = pd.DataFrame(X)
X.head()
```

```python
#Label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = le.fit_transform(X)
X = pd.DataFrame(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.5)
```

```python
#import model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```
